```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("hypothetical data.xlsx",sheet_name="static")
df.head()
```

```python
def mean_squared_error(y, x1, x2, x3, x4):
    y_pred = x1*x2
    loss = np.mean((y - y_pred)**2)
    return loss

def loss_calculator(df,a): #update loss column
    df['Loss'] = df.apply(lambda row: mean_squared_error(row['y'], a[0]*row['x1'], a[1]*row['x2'], a[2]*row['x3'], a[3]*row['x4']), axis=1)
    return df

def performance_calculator(sampled_df,a,i):
    a[i] = 0
    loss_calculator(sampled_df,a)
    mean = sampled_df['Loss'].mean()
    a[i] = 1
    return mean

def update_df_itr(df_itr,new_p,k):
    new_itr = [k] + new_p
    # convert the list to a pandas Series or dictionary
    new_row = pd.Series(new_itr, index=df_itr.columns)
    # append the new row to the dataframe
    df_itr = df_itr.append(new_row, ignore_index=True)
    return df_itr

loss_calculator(df,a=[1,1,1,1])
df.head()
```

```python
# Create a dataframe using the dictionary
df_itr = pd.DataFrame({'iteration':[1],'p1':[0.5],'p2':[0.5],'p3':[0.5],'p4':[0.5],})
df_itr.head()
```

```python
#given branch x1=0
a = [1,1,1,1]
p = [0.5,0.5,0.5,0.5]
k_max = 100

for k in range(0,k_max):
    sampled_df = df.sample(n=50)
    for i in range(0,len(a)):
        for j in range(0,len(a)):
            performance_i_out = performance_calculator(sampled_df,a,i)
            performance_j_out = performance_calculator(sampled_df,a,j)
            if performance_i_out >= performance_j_out:
                p[i] *= 1.001
                p[j] /= 1.001
    print(k,p)

```
